```python
# Complete Matplotlib Graphs Reference
# A comprehensive collection of all essential matplotlib graphs

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Wedge
import seaborn as sns

# Set style for better looking plots
plt.style.use('default')
sns.set_palette("husl")

# Create sample data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]
data = np.random.randn(100)
```

```python
# 1. LINE PLOT
print("\n1. LINE PLOT")
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', linewidth=2)
plt.plot(x, y2, label='cos(x)', linewidth=2)
plt.title('Line Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 2. SCATTER PLOT
print("\n2. SCATTER PLOT")
plt.figure(figsize=(10, 6))
x_scatter = np.random.randn(50)
y_scatter = 2 * x_scatter + np.random.randn(50)
plt.scatter(x_scatter, y_scatter, alpha=0.7, s=60)
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 3. BAR CHART
print("\n3. BAR CHART")
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# 4. HORIZONTAL BAR CHART
print("\n4. HORIZONTAL BAR CHART")
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# 5. HISTOGRAM
print("\n5. HISTOGRAM")
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. PIE CHART
print("\n6. PIE CHART")
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.show()
```

```python
# 7. BOX PLOT
print("\n7. BOX PLOT")
plt.figure(figsize=(10, 6))
data_box = [np.random.randn(100) + i for i in range(5)]
plt.boxplot(data_box, labels=categories)
plt.title('Box Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# 8. VIOLIN PLOT
print("\n8. VIOLIN PLOT")
plt.figure(figsize=(10, 6))
plt.violinplot(data_box, positions=range(1, 6))
plt.xticks(range(1, 6), categories)
plt.title('Violin Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# 9. AREA PLOT
print("\n9. AREA PLOT")
plt.figure(figsize=(10, 6))
plt.fill_between(x, y1, alpha=0.5, label='sin(x)')
plt.fill_between(x, y2, alpha=0.5, label='cos(x)')
plt.title('Area Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.show()
```

```python
# 10. STEM PLOT
print("\n10. STEM PLOT")
plt.figure(figsize=(10, 6))
x_stem = np.arange(10)
y_stem = np.random.randn(10)
plt.stem(x_stem, y_stem)
plt.title('Stem Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 11. STEP PLOT
print("\n11. STEP PLOT")
plt.figure(figsize=(10, 6))
plt.step(x[:20], y1[:20], where='mid', linewidth=2)
plt.title('Step Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 12. ERROR BAR PLOT
print("\n12. ERROR BAR PLOT")
plt.figure(figsize=(10, 6))
x_err = np.arange(5)
y_err = [20, 35, 30, 35, 27]
yerr = [2, 3, 4, 1, 2]
plt.errorbar(x_err, y_err, yerr=yerr, fmt='o-', capsize=5)
plt.title('Error Bar Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 13. CONTOUR PLOT
print("\n13. CONTOUR PLOT")
plt.figure(figsize=(10, 8))
X = np.linspace(-3, 3, 100)
Y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contour(X, Y, Z, levels=10)
plt.colorbar()
plt.title('Contour Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 14. CONTOUR FILLED PLOT
print("\n14. CONTOUR FILLED PLOT")
plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.title('Filled Contour Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 15. HEATMAP
print("\n15. HEATMAP")
plt.figure(figsize=(10, 8))
data_heat = np.random.randn(10, 10)
plt.imshow(data_heat, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()
```

```python
# 16. 3D SURFACE PLOT
print("\n16. 3D SURFACE PLOT")
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot Example')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_zlabel('Z values')
plt.show()
```

```python
# 17. 3D SCATTER PLOT
print("\n17. 3D SCATTER PLOT")
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
x_3d = np.random.randn(100)
y_3d = np.random.randn(100)
z_3d = np.random.randn(100)
ax.scatter(x_3d, y_3d, z_3d, c=z_3d, cmap='viridis')
ax.set_title('3D Scatter Plot Example')
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_zlabel('Z values')
plt.show()
```

```python
# 18. POLAR PLOT
print("\n18. POLAR PLOT")
plt.figure(figsize=(8, 8))
theta = np.linspace(0, 2*np.pi, 100)
r = 1 + np.sin(3*theta)
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_title('Polar Plot Example')
plt.show()
```

```python
# 19. QUIVER PLOT (VECTOR FIELD)
print("\n19. QUIVER PLOT")
plt.figure(figsize=(10, 8))
x_q = np.arange(-3, 4)
y_q = np.arange(-3, 4)
X_q, Y_q = np.meshgrid(x_q, y_q)
U = -Y_q
V = X_q
plt.quiver(X_q, Y_q, U, V)
plt.title('Quiver Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 20. STREAMPLOT
print("\n20. STREAMPLOT")
plt.figure(figsize=(10, 8))
plt.streamplot(X_q, Y_q, U, V, density=2.5)
plt.title('Stream Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 21. HEXBIN PLOT
print("\n21. HEXBIN PLOT")
plt.figure(figsize=(10, 8))
x_hex = np.random.randn(1000)
y_hex = np.random.randn(1000)
plt.hexbin(x_hex, y_hex, gridsize=20, cmap='Blues')
plt.colorbar()
plt.title('Hexbin Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# 22. STACKED BAR CHART
print("\n22. STACKED BAR CHART")
plt.figure(figsize=(10, 6))
values1 = [20, 35, 30, 35, 27]
values2 = [25, 30, 15, 30, 22]
plt.bar(categories, values1, label='Series 1')
plt.bar(categories, values2, bottom=values1, label='Series 2')
plt.title('Stacked Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# 23. GROUPED BAR CHART
print("\n23. GROUPED BAR CHART")
plt.figure(figsize=(10, 6))
x_pos = np.arange(len(categories))
width = 0.35
plt.bar(x_pos - width/2, values1, width, label='Series 1')
plt.bar(x_pos + width/2, values2, width, label='Series 2')
plt.title('Grouped Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x_pos, categories)
plt.legend()
plt.show()
```

```python
# 25. DONUT CHART
print("\n25. DONUT CHART")
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(values, labels=categories, autopct='%1.1f%%', 
                                startangle=90, pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart Example')
plt.show()
```

```python
# 26. WATERFALL CHART (CUSTOM)
print("\n26. WATERFALL CHART")
plt.figure(figsize=(10, 6))
values_waterfall = [100, -30, 20, -10, 15, -25]
labels_waterfall = ['Start', 'Dec 1', 'Inc 1', 'Dec 2', 'Inc 2', 'End']
cumulative = np.cumsum([0] + values_waterfall[:-1])
colors = ['green' if v > 0 else 'red' for v in values_waterfall]
colors[0] = 'blue'  # Start
colors[-1] = 'blue'  # End

for i, (label, value, cum) in enumerate(zip(labels_waterfall, values_waterfall, [0] + list(cumulative[1:]))):
    plt.bar(i, abs(value), bottom=cum if value > 0 else cum + value, color=colors[i], alpha=0.7)
    plt.text(i, cum + value/2, f'{value:+}', ha='center', va='center')

plt.title('Waterfall Chart Example')
plt.xticks(range(len(labels_waterfall)), labels_waterfall)
plt.ylabel('Values')
plt.show()
```
