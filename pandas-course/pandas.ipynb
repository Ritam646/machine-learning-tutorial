```python
data4 = {
    "name": ["Arnab",None, "Sourav", "Sayan", "Rohit", "Rahul", "Virat"],
    "age": [25, None, 30, 22, 28, 26, 29],
    "city": ["Kolkata", None,"Delhi", "Mumbai", "Chennai", "Bangalore", "Hyderabad"],
    "salary": [50000,None, 60000, 55000, 70000, 65000, 80000],
    "performance": [85,None, 90, 88, 92, 87, 95]
}
```

```python
new_df1 = pd.DataFrame(data4)
print(new_df1)
```

```python
# new_df1.fillna(0, inplace=True)  # Fill missing values with 0
print(new_df1)
```

```python
new_df1['age'].fillna(new_df1['age'].mean(), inplace=True)  # Fill missing values with the mean of the column
new_df1['salary'].fillna(new_df1['salary'].median(), inplace=True)  # Fill missing values with the median of the column
new_df1['city'].fillna('Unknown', inplace=True)  # Fill missing values with a specific value
new_df1['performance'].fillna(new_df1['performance'].mode()[0], inplace=True)  # Fill missing values with the mode of the column
new_df1['name'].fillna('Unknown', inplace=True)  # Fill missing values with a specific value
print(new_df1)
```

```python
new_df2.interpolate(method='linear', inplace=True)  # Interpolate missing values using linear interpolation
print(new_df2)
```

```python
# Sorting the DataFrame
new_df2.sort_values(by='salary', ascending=False, inplace=True)  # Sort by salary in descending order
print(new_df2) 
new_df2.sort_values(by='salary', ascending=True, inplace=True)  # Sort by salary in ascending order
print(new_df2)
```

```python
new_df2.sort_values(by=['salary', 'age'], ascending=[True, False], inplace=True)  # Sort by multiple columns
print(new_df2)
```

```python
#summary statistics
summary_stats = new_df2.describe()  # Get summary statistics of the DataFrame
print(summary_stats)
mean_salary = new_df2['salary'].mean()  # Calculate the mean of the salary column
print("Mean Salary:", mean_salary)  
median_salary = new_df2['salary'].median()  # Calculate the median of the salary column
print("Median Salary:", median_salary)
# Calculate the mode of the salary column
mode_salary = new_df2['salary'].mode()[0]  # Get the first mode 
print("Mode Salary:", mode_salary) 
max_salary = new_df2['salary'].max()  # Get the maximum salary
print("Max Salary:", max_salary)
min_salary = new_df2['salary'].min()  # Get the minimum salary
print("Min Salary:", min_salary)
```

```python
#graouping the dataframe
grouped_df = new_df2.groupby('age')["salary"].sum()  # Group by age and calculate the sum of salaries
print(grouped_df)
#groupby functions
grouped_df_mean = new_df2.groupby('age')["salary"].mean()  # Group by age and calculate the mean of salaries
print(grouped_df_mean)
grouped_df_median = new_df2.groupby('age')["salary"].median()  # Group by age and calculate the median of salaries
print(grouped_df_median)
grouped_df_max = new_df2.groupby('age')["salary"].max()  # Group by age and calculate the maximum salary
print(grouped_df_max)
grouped_df_min = new_df2.groupby('age')["salary"].min()  # Group by age and calculate the minimum salary
print(grouped_df_min)
grouped_df_std = new_df2.groupby('age')["salary"].std()  # Group by age and calculate the standard deviation of salaries
print(grouped_df_std)
# Group by age and calculate the count of salaries
grouped_df_count = new_df2.groupby('age')["salary"].count()
print(grouped_df_count)
# Group by multiple columns and calculate the sum of salaries
grouped_df_multi = new_df2.groupby(['age', 'city'])["salary"].sum()
print(grouped_df_multi)
```

```python
#pivoting the dataframe
pivot_df = new_df2.pivot_table(index='age', columns='city', values='salary', aggfunc='mean')  # Create a pivot table with age as index, city as columns, and mean salary as values
print(pivot_df)
```

```python
#merging dataframes
customer_data = {
    "customer_id": [1, 2, 3, 4, 5],
    "customer_name": ["Arnab", "Sourav", "Sayan", "Rohit", "Rahul"]
}   
customer_df = pd.DataFrame(customer_data)
print(customer_df)
order_data = {
    "order_id": [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 3, 5, 6],
    "order_amount": [2500, 3000, 1500, 4000, 3500]
}   
order_df = pd.DataFrame(order_data)
print(order_df)
merged_df = pd.merge(customer_df, order_df, on="customer_id",how="inner")  # Merge the customer and order DataFrames on customer_id using an inner join
print(merged_df)
merged_df_outer = pd.merge(customer_df, order_df, on="customer_id", how="outer")  # Merge the customer and order DataFrames on customer_id using an outer join
print(merged_df_outer)
#left join
merged_df_left = pd.merge(customer_df, order_df, on="customer_id", how="left")  # Merge the customer and order DataFrames on customer_id using a left join
print(merged_df_left)
#right join
merged_df_right = pd.merge(customer_df, order_df, on="customer_id", how="right")  # Merge the customer and order DataFrames on customer_id using a right join
print(merged_df_right)
#merging dataframes with multiple keys
customer_data1 = {
    "order_id": [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 3, 4, 5],
    "customer_name": ["Arnab", "Sourav", "Sayan", "Rohit", "Rahul"]
}   
customer_df1 = pd.DataFrame(customer_data1)
print(customer_df1)
order_data1 = {
    "order_id": [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 3, 6, 5]
}   
order_df1 = pd.DataFrame(order_data1)
print(order_df1)
merged_df1 = pd.merge(customer_df1, order_df1, on=["customer_id", "order_id"], how="inner")  # Merge the customer and order DataFrames on customer_id and order_id using an inner join
print(merged_df1)
#merging dataframes with multiple keys using left join
merged_df_left = pd.merge(customer_df1, order_df1, on=["customer_id", "order_id"], how="left")  # Merge the customer and order DataFrames on customer_id and order_id using a left join
print(merged_df_left)

```

```python
#concatenation 
df_region1 = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'region': ['East', 'East', 'West', 'West']
})

df_region2 = pd.DataFrame({
    'customer_id': [4, 5, 6, 7],
    'region': ['North', 'North', 'South', 'South']
})

df_concat = pd.concat([df_region1, df_region2],axis=0, ignore_index=True)
print(df_concat)
#horizontal concatenation
df_concat = pd.concat([df_region1, df_region2],axis=1)
print(df_concat)
```
