```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
def transform_family_size(num):
    if num == 1:
        return 'alone'
    elif(num>1 and num<=4):
        return 'small'
    else:
        return 'large'
```

```python
df['Family_Size'].apply(transform_family_size)

# or df['Family_Size'].apply(lambda x: transform_family_size(x))
```
